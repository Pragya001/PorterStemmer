```python
class Porter_Stemmer:
    def isAEIOU(self, letter):
        
        if letter == 'a' or letter == 'e' or letter == 'i' or letter == 'o' or letter == 'u':
            return True
        else:
            return False
    
    def isVowel(self, word):
    
        return not self.isConsonant(word, i)
    
    def isConsonant(self, word, i):
        
        if self.isAEIOU(word[i]):
            return False
        elif word[i] == 'y' and not self.isAEIOU(word[i-1]):
            return False
        else:
            return True
    
    def replace(self, word, find, replace):
        
        index = word.rfind(find)
        start = word[:index]
        new_str = start + replace
        
        return new_str
    
    def containsVowel(self, word):
        
        for i in range(len(word)):
            if self.isAEIOU(word[i]):
                return True
        return False
    
    def countM(self, word):
        
        forms = self.getForms(word)
        m = forms.count('VC')
        return m    
    
    def getForms(self, word):
        
        forms = []
        string_form = ""
        
        for i in range(len(word)):
            if self.isConsonant(word, i):
                if i != 0:
                    if forms[-1] != 'C':
                        forms.append('C')
                else:
                    forms.append('C')
            else:
                if i != 0:
                    if forms[-1] != 'V':
                        forms.append('V')
                else:
                    forms.append('V')
        
        for i in forms:
            string_form += i
        
        return string_form
    
    def doubleCons(self, word):
        
        if len(word) >= 2:
            if self.isConsonant(word, -1) and self.isConsonant(word, -1) and word[-1] == word[-2]:
                return True
            else:
                return False
        else:
            return False
    
    def getcvc(self, word):
        
        forms = self.getForms(word)
        if forms.endswith('CVC') and word[-1] != 'w' and word[-1] != 'x' and word[-1] != 'y':
                return True
        else:
            return False
    
    def check_m0(self, word, suff):
        
        index = word.rfind(suff)
        stem = word[:index]
        if self.countM(stem) > 0:
            return True
        else:
            return False
    
    def check_m1(self, word, suff):
        
        index = word.rfind(suff)
        stem = word[:index]
        if self.countM(stem) > 1:
            return True
        else:
            return False
    
    def step1a(self, word):
        
        if word.endswith('sses'):
            word = self.replace(word, 'sses', 'ss')
        elif word.endswith('ies'):
            word = self.replace(word, 'ies', 'i')
        elif word.endswith('ss'):
            word = self.replace(word, 'ss', 'ss')
        elif word.endswith('s'):
            word = self.replace(word, 's', '')
        else:
            pass
        return word
    
    def step1b(self, word):
        
        flag = 0
        if word.endswith('eed'):
            index = word.rfind('eed')
            stem = word[:index]
            if self.countM(stem) > 0:
                word = self.replace(word, 'eed', 'ee')
                
        elif word.endswith('ed'):
            index = word.rfind('ed')
            stem = word[:index]
            if self.containsVowel(stem):
                word = self.replace(word, 'ed', '')
                flag = 1
                
        elif word.endswith('ing'):
            index = word.rfind('ing')
            stem = word[:index]
            if self.containsVowel(stem):
                word = self.replace(word, 'ing', '')
                flag = 1
                
        if flag == 1:
            if word.endswith('at') or word.endswith('bl') or word.endswith('iz'):
                word += 'e'
            elif self.doubleCons(word) and not word.endswith('l') and not word.endswith('s') and not word.endswith('z'):
                word = word[:-1]
            elif self.countM(word) == 1 and self.getcvc(word):
                word += 'e'
            else:
                pass
        else: 
            pass
        return word
    
    def step1c(self, word):
        
        if word.endswith('y') and self.containsVowel(word[:-1]):
            word = word[:-1]
            word += 'i'
        return word
    
    def step2(self, word):
        
        if word.endswith('ational'):
            if self.check_m0(word, 'ational'):
                word = self.replace(word, 'ational', 'ate')
        elif word.endswith('tional'):
            if self.check_m0(word, 'tional'):
                word = self.replace(word, 'tional', 'tion')
        elif word.endswith('enci'):
            if self.check_m0(word, 'enci'):
                word = self.replace(word, 'enci', 'ence')
        elif word.endswith('anci'):
            if self.check_m0(word, 'anci'):
                word = self.replace(word, 'anci', 'ance')
        elif word.endswith('izer'):
            if self.check_m0(word, 'izer'):
                word = self.replace(word, 'izer', 'ize')
        elif word.endswith('abli'):
            if self.check_m0(word, 'abli'):
                word = self.replace(word, 'abli', 'able')
        elif word.endswith('alli'):
            if self.check_m0(word, 'alli'):
                word = self.replace(word, 'alli', 'al')
        elif word.endswith('entli'):
            if self.check_m0(word, 'entli'):
                word = self.replace(word, 'entli', 'ent')
        elif word.endswith('abli'):
            if self.check_m0(word, 'abli'):
                word = self.replace(word, 'abli', 'able')
        elif word.endswith('eli'):
            if self.check_m0(word, 'eli'):
                word = self.replace(word, 'eli', 'e')
        elif word.endswith('ousli'):
            if self.check_m0(word, 'ousli'):
                word = self.replace(word, 'ousli', 'ous')
        elif word.endswith('ization'):
            if self.check_m0(word, 'ization'):
                word = self.replace(word, 'ization', 'ize')
        elif word.endswith('ation'):
            if self.check_m0(word, 'ation'):
                word = self.replace(word, 'ation', 'ate')
        elif word.endswith('ator'):
            if self.check_m0(word, 'ator'):
                word = self.replace(word, 'ator', 'ate')
        elif word.endswith('alism'):
            if self.check_m0(word, 'alism'):
                word = self.replace(word, 'alism', 'al')
        elif word.endswith('iveness'):
            if self.check_m0(word, 'iveness'):
                word = self.replace(word, 'iveness', 'ive')
        elif word.endswith('fulness'):
            if self.check_m0(word, 'fulness'):
                word = self.replace(word, 'fulness', 'ful')
        elif word.endswith('ousness'):
            if self.check_m0(word, 'ousness'):
                word = self.replace(word, 'ousness', 'ous')
        elif word.endswith('aliti'):
            if self.check_m0(word, 'aliti'):
                word = self.replace(word, 'aliti', 'al')
        elif word.endswith('iviti'):
            if self.check_m0(word, 'iviti'):
                word = self.replace(word, 'iviti', 'ive')
        elif word.endswith('biliti'):
            if self.check_m0(word, 'biliti'):
                word = self.replace(word, 'biliti', 'ble')
        else:
            pass
        return word
                
    def step3(self, word):
        
        if word.endswith('icate'):
            if self.check_m0(word, 'icate'):
                word = self.replace(word, 'icate', 'ic')
        elif word.endswith('ative'):
            if self.check_m0(word, 'ative'):
                word = self.replace(word, 'ative', '')        
        elif word.endswith('alize'):
            if self.check_m0(word, 'alize'):
                word = self.replace(word, 'alize', 'al')        
        elif word.endswith('iciti'):
            if self.check_m0(word, 'iciti'):
                word = self.replace(word, 'iciti', 'ic')        
        elif word.endswith('ical'):
            if self.check_m0(word, 'ical'):
                word = self.replace(word, 'ical', 'ic')        
        elif word.endswith('ful'):
            if self.check_m0(word, 'ful'):
                word = self.replace(word, 'ful', '')        
        elif word.endswith('ness'):
            if self.check_m0(word, 'ness'):
                word = self.replace(word, 'ness', '')        
        else:
            pass
        return word
    
    def step4(self, word):
        
        if word.endswith('al'):
            if self.check_m1(word, 'al'):
                word = self.replace(word, 'al', '')
        elif word.endswith('ance'):
            if self.check_m1(word, 'ance'):
                word = self.replace(word, 'ance', '')        
        elif word.endswith('ence'):
            if self.check_m1(word, 'ence'):
                word = self.replace(word, 'ence', '')        
        elif word.endswith('er'):
            if self.check_m1(word, 'er'):
                word = self.replace(word, 'er', '')        
        elif word.endswith('ic'):
            if self.check_m1(word, 'ic'):
                word = self.replace(word, 'ic', '')        
        elif word.endswith('able'):
            if self.check_m1(word, 'able'):
                word = self.replace(word, 'able', '')        
        elif word.endswith('ible'):
            if self.check_m1(word, 'ible'):
                word = self.replace(word, 'ible', '')        
        elif word.endswith('ant'):
            if self.check_m1(word, 'ant'):
                word = self.replace(word, 'ant', '')        
        elif word.endswith('ement'):
            if self.check_m1(word, 'ement'):
                word = self.replace(word, 'ement', '')        
        elif word.endswith('ment'):
            if self.check_m1(word, 'ment'):
                word = self.replace(word, 'ment', '')        
        elif word.endswith('ent'):
            if self.check_m1(word, 'ent'):
                word = self.replace(word, 'ent', '')        
        elif word.endswith('ous'):
            if self.check_m1(word, 'ous'):
                word = self.replace(word, 'ous', '')        
        elif word.endswith('ism'):
            if self.check_m1(word, 'ism'):
                word = self.replace(word, 'ism', '')        
        elif word.endswith('ate'):
            if self.check_m1(word, 'ate'):
                word = self.replace(word, 'ate', '')        
        elif word.endswith('iti'):
            if self.check_m1(word, 'iti'):
                word = self.replace(word, 'iti', '')        
        elif word.endswith('ou'):
            if self.check_m1(word, 'ou'):
                word = self.replace(word, 'ou', '')        
        elif word.endswith('ive'):
            if self.check_m1(word, 'ive'):
                word = self.replace(word, 'ive', '')        
        elif word.endswith('ize'):
            if self.check_m1(word, 'ize'):
                word = self.replace(word, 'ize', '')        
        elif word.endswith('ion'):
            index = word.rfind('ion')
            stem = word[:index]
            if self.countM(stem) > 1 and (stem.endswith('s') or stem.endswith('t')):
                word = stem
        else:
            pass
        return word
        
    def step5a(self, word):
        
        if word.endswith('e'):
            if self.check_m1(word, 'e'):
                word = word[:-1]
        elif word.endswith('e'):
            stem = word[:-1]
            if self.countM(stem) == 1 and not getCVC(stem):
                word = word[:-1]
        else:
            pass
        return word
            
    def step5b(self, word):
        
        if self.countM(word) > 1 and self.doubleCons(word) and word.endswith('l'):
            word = word[:-1]
        return word
```

```python
def processing(word):
    ps = Porter_Stemmer()
    word = word.lower()
    word = ps.step1a(word)
    word = ps.step1b(word)
    word = ps.step1c(word)
    word = ps.step2(word)
    word = ps.step3(word)
    word = ps.step4(word)
    word = ps.step5a(word)
    word = ps.step5b(word)
    
    return word
```
